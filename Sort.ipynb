# Реализация сортировки вставками в Python
## Алгоритм сортировки вставками работает точно так же, как в примере с колодой карт. Вот реализация на Python:

В отличие от пузырьковой сортировки, эта реализация сортировки вставками создает отсортированный список, сдвигая меньшие элементы влево. Разберем insertion_sort() построчно:

- Строка 2 устанавливает цикл, который определяет key_item, что функция будет позиционировать во время каждой итерации. Обратите внимание, что цикл начинается со второго элемента в списке и продолжается до последнего элемента.

- Строка 3 инициализируется key_item элементом, который функция пытается разместить.

- Строка 4 инициализирует переменную, которая будет последовательно указывать на каждый элемент слева от key item. Это элементы, которые будут последовательно сравниваться с key_item.

- Строка 5 сравнивает key_item каждое значение слева от нее, используя while цикл, сдвигая элементы, чтобы освободить место для размещения key_item.

- Строка 9 занимает правильное место key_item после того, как алгоритм сдвинет все большие значения вправо.

Теперь вот краткое изложение шагов алгоритма при сортировке массива:

1. Алгоритм начинается key_item = 2 и проходит через подмассив слева от него, чтобы найти для него правильную позицию. В данном случае это подмассив [8].

2. Так 2 < 8 как алгоритм сдвигает элемент 8 на одну позицию вправо. Результирующий массив в этот момент равен [8, 8, 6, 4, 5].

3. Поскольку в подмассиве больше нет элементов, key_item теперь он помещается в новое положение, и конечный массив равен [2, 8, 6, 4, 5].

4. Второй проход начинается key_item = 6 и проходит через подмассив, расположенный слева от него, в данном случае [2, 8].

5. Так 6 < 8 как алгоритм сдвигает 8 вправо. Результирующий массив в этот момент равен [2, 8, 8, 4, 5].

6. Так 6 > 2 как , алгоритму не нужно продолжать обход подмассива, поэтому он позиционирует key_itemи завершает второй проход. На данный момент результирующий массив равен [2, 6, 8, 4, 5].

7. Третий проход по списку помещает элемент 4 в правильное положение, а четвертый проход помещает элемент 5 в правильное место, оставляя массив отсортированным.



```python
def insertion_sort(array):
    for i in range(1, len(array)):
        key_item = array[i]
        j = i - 1
        while j >= 0 and array[j] > key_item:
            array[j + 1] = array[j]
            j -= 1

        array[j + 1] = key_item

    return array

arr = [8, 2, 6, 4, 5]
insertion_sort(arr)
```
